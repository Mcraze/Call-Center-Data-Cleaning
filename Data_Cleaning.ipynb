```python
#Importing necessary Libraries
import pandas as pd
```

```python
#Creating dataframe
df = pd.read_excel('Customer Call List.xlsx')
```

```python
#Reading data from dataframe
df
```

```python
#Dropping duplicate values from the dataframe
df = df.drop_duplicates()

df
```

```python
#Removing column which is not required
df = df.drop(columns = 'Not_Useful_Column')

df
```

```python
#There are some errors which are visible in the Last_Name column
#Use lstrip() for left side removal and rstrip() for right side
#You can use individual strip too: df['Last_Name'] = df['Last_Name'].str.strip('/')

#Specify the values which needs to be removed
df['Last_Name'] = df['Last_Name'].str.strip('._/')

df
```

```python
#Cleaning Phone Number. Firstly remove the non numeric values such as - / | etc
#Here we are specifing values which should not be removed basically remove values except a-zA-z0-9
#a-zA-z0-9 because there are NaN and Na values available and we don't want them to be replaced
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-z0-9]','')

df
```

```python
#Before adding - inside the phone numbers we need to convert them to string in order to use lambda function
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
#Now applying changes in the phone number and adding - between numbers at 3 intervals
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df
```

```python
#Removing the nan-- and na-- values from the dataframe
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')

df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')

df
```

```python
#Cleaning Addresses for increasing readability

#Splitting Columns into Street_Address, State, Zipcode

df[['Street_Address', 'City_Name', 'Zipcode']] = df['Address'].str.split(',', 2 , expand = True)

df.head()
```

```python
#Replace Yes and No to Y and N in the Paying Customers column
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

df
```

```python
#Let us do the same for Do_Not_Contact column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

df
```

```python
#Replacing N/A values in the dataframe with blanks using fillna function
df.fillna('')
```

```python
#Now we only want a list of people who we are allowed to make a call
#Remove the rows which says yes to do not call
df1 = df

for x in df1.index:
  if df1.loc[x , 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace = True)

df1
```

```python
#Lets fill the NaN with N now in the Do Not Call column
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].fillna('N')

df1
```

```python
#Let us remove rows which do not have any phone number

for x in df1.index:
  if df1.loc[x, 'Phone_Number'] == '':
    df1.drop(x, inplace = True)

df1
```

```python
#Resetting index for better readability
df1 = df1.reset_index(drop = True)
```

```python
#Final Product of our Data Cleaning
df1
```

```python
#Export to an excel file for further needs

df1.to_excel('sample_data/Cleaned Data.xlsx')
```
